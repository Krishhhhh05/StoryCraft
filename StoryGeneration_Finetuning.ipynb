```python
from huggingface_hub import HfApi, HfFolder, Repository
import os
import shutil

# Parameters
model_directory = '/content/drive/MyDrive/LLaMA-Factory/saves/Mistral-7B-Chat/lora/train_2024-03-11-16-49-33'  # Path to your model directory
repo_name = "Story_Generation"  # Choose a name for your Hugging Face repository
organization = None  # Specify an organization if you want to create the repo there, else it will be created under your user
private_repo = False  # Set to True if you want to create a private repository

# Hugging Face authentication
hf_token = "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"  # Replace with your Hugging Face token
HfFolder.save_token(hf_token)  # Save token for subsequent API calls

# Initialize HfApi for user information
api = HfApi()
user = api.whoami(token=hf_token)
username = user["name"] if organization is None else organization
repo_id = f"{username}/{repo_name}"

# Create a repository on Hugging Face
api.create_repo(repo_id=repo_id, token=hf_token, private=private_repo, exist_ok=True)

# Clone the newly created repository to a local directory
repo_local_path = os.path.join("/tmp", repo_name)
repo = Repository(repo_local_path, clone_from=repo_id, use_auth_token=hf_token)

# Copy model files from the model_directory to the cloned repository directory
# You might want to filter which files to copy depending on your model structure and what you want to upload
for filename in os.listdir(model_directory):
    src_file = os.path.join(model_directory, filename)
    dst_file = os.path.join(repo_local_path, filename)
    if os.path.isdir(src_file):
        shutil.copytree(src_file, dst_file)
    else:
        shutil.copy2(src_file, dst_file)

# Add, commit, and push files to the repository
repo.git_add(auto_lfs_track=True)  # Automatically track large files with Git LFS
repo.git_commit("Initial commit with the fine-tuned model")
repo.git_push()

print(f"Model successfully uploaded to: https://huggingface.co/{repo_id}")
```
